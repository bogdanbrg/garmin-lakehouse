## Silver Layer — Gear Statistics
Cleans and standardizes gear usage statistics from Garmin API.

**Transformations:**
* Convert Unix timestamps (milliseconds) → dates
* Distance: meters → kilometers
* Standardize column names (uuid → gear_id)
* Round numeric values

```python
%sql
CREATE OR REPLACE TABLE garmin_lakehouse.silver.gear_stats AS
SELECT
    uuid as gear_id,  -- Match column name from gear_list
    
    -- Convert Unix timestamps (milliseconds) to readable dates
    CAST(FROM_UNIXTIME(CAST(updateDate AS BIGINT) / 1000) AS DATE) as updated_at,
    
    -- Convert distance from meters to kilometers
    ROUND(totalDistance / 1000, 2) as total_distance_km,
    
    ROUND(totalActivities, 0) as total_activities,
    
    -- Metadata
    CURRENT_TIMESTAMP() as loaded_at
    
FROM garmin_lakehouse.raw.gear_stats
```
