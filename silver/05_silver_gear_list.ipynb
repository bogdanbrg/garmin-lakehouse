## Silver Layer — Gear List
Cleans and standardizes gear inventory data from Garmin API.

**Transformations:**
* Drop image columns (void types)
* Standardize column names (uuid → gear_id)
* Convert dates to proper format
* Convert maximum distance to kilometers
* Select only active and relevant columns

```python
%sql
CREATE OR REPLACE TABLE garmin_lakehouse.silver.gear_list AS
SELECT
    uuid as gear_id,
    gearPk as gear_pk,
    userProfilePk as user_profile_pk,
    
    -- Gear identification
    customMakeModel as gear_name,
    gearTypeName as gear_type,
    gearStatusName as gear_status,
    
    -- Distance limit (convert meters to km)
    ROUND(maximumMeters / 1000, 0) as max_distance_km,
    
    -- Dates (parse from ISO format)
    CAST(dateBegin AS DATE) as date_start,
    CAST(dateEnd AS DATE) as date_end,
    
    -- Flags
    notified,
    
    -- Metadata
    CURRENT_TIMESTAMP() as loaded_at
    
FROM garmin_lakehouse.raw.gear_list
-- Exclude the void image columns: imageNameLarge, imageNameMedium, imageNameSmall
```

```python
%sql
SELECT *
FROM garmin_lakehouse.silver.gear_list
ORDER BY date_start DESC
```
