## Silver Layer — Activity Gear
Clean bridge table mapping activities to gear.

**Purpose:** One row per activity showing which gear was used (many-to-one relationship).

**Transformations:**
* Standardize column names (uuid → gear_id)
* Cast activity ID to string for consistency
* Drop void image columns from bronze
* Preserve join keys: activity_id, gear_pk

```python
%sql
CREATE OR REPLACE TABLE garmin_lakehouse.silver.activity_gear AS
SELECT
    CAST(activityId AS STRING) as activity_id,
    uuid as gear_id,  -- Match naming from gear_list
    gearPk as gear_pk,
    userProfilePk as user_profile_pk,
    customMakeModel as gear_name,
    
    -- Metadata
    CURRENT_TIMESTAMP() as loaded_at
    
FROM garmin_lakehouse.raw.activity_gear
```

```python
%sql
SELECT *
FROM garmin_lakehouse.silver.activity_gear
ORDER BY loaded_at DESC
LIMIT 20
```
